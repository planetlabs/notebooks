# Tile Services Mapping Demo with Leaflet

This notebook demonstrates how to render Planet Mosaics via [XYZ Tile Services](https://docs.planet.com/develop/apis/tiles/#basemap-tile-service).

In this example, you will build an interactive map that looks like this one:

![leaflet_map.png](images/leaflet_map.png)

## Setup

```python
import os
from ipyleaflet import TileLayer, Map, basemaps, LayersControl
```

```python
# setup Planet API key
pl_api_key = os.getenv('PL_API_KEY')
if not pl_api_key:
    import getpass
    pl_api_key = getpass.getpass('Planet API Key: ')
```

```python
# You may need to adjust this to a location within your own Area of Access
lat, lon = 37.7749, -122.4194
```

## Make the Map

```python
m = Map(
    center=(lat, lon),
    zoom=12,
    basemap=basemaps.CartoDB.Positron # this is an underlying basemap that will show up below our Planet tile layers
)

# replace these with different mosaic names if you do not have access to global monthly mosaics
mosaic_names = ['global_monthly_2024_11_mosaic', 'global_monthly_2024_08_mosaic', 'global_monthly_2024_05_mosaic']
for mosaic in mosaic_names:
    layer = TileLayer(
        url=f'https://tiles.planet.com/basemaps/v1/planet-tiles/{mosaic}/gmap/{{z}}/{{x}}/{{y}}.png?api_key={pl_api_key}',
        name=mosaic,
        attribution='Planet Labs PBC'
    )
    m.add_layer(layer)

# Add layer control
layer_control = LayersControl(position='topright')
m.add_control(layer_control)

# Display the map
m
```

You should now see a rendered map with one layer per mosaic.
Try toggling layers on/off to see how the area in view changed over time.
